#  The Goal of Collecting this Dataset

The goal of our project is to build a model that can predict the adaptivity level of a student with a high degree of accuracy. We will be using different data mining tasks to achieve this goal.

## Dataset Source
The dataset was obtained from Kaggle.<br>
https://www.kaggle.com/datasets/mdmahmudulhasansuzan/students-adaptability-level-in-online-education

## General information:
Our dataset has 14 attributes and 1,205 objects.
<br>The class label is "Adaptivity Level".
